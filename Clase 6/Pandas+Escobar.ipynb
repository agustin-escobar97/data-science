```python
import urllib.request
import pandas as pd
```

```python
url = "https://drive.google.com/uc?id=19G9QAuL2py_ucA3RixZ1QPTkEz_I5u3W&authuser=0&export=download" #importamos el dataframe desde la url
df=pd.read_csv(url, encoding="latin-1")#asignamos el dataframe a df
df.dropna(inplace=True) #les sacamos las filas en blanco y mantenemos los cambios
consignas = df #construimos el objeto a partir del df original
consignas["Year"] = consignas["Year"].astype(int) #por alguna razon los años los toma como float asi que lo pasamos a enteros
```

```python
#Consigna 1: con el archivo brindado, construir un objeto de series de tiempo con el indioce igual al año de los juegos olimpicos

consignas["Year"] = pd.to_datetime(consignas["Year"], format="%Y")#construimos la serie de tiempo
consignas.set_index("Year")#y convertimos el año en el indice del dataframe
```

```python
#Consigna 2: Separar los nombres de apellidos en dos columnas distintas usando la coma como separador

nombre_apellido = consignas["Athlete"].str.split(", ")#separamos el nombre del apellido, usando como separador la coma
consignas["nombre"]=nombre_apellido.str[1]#creamos la columna nombre y le pasamos como dato el segundo split
consignas["apellido"]=nombre_apellido.str[0]#creamos la columna apellido y le pasamos como dato el primer split
consignas[["nombre", "apellido"]]#aqui traeremos solamente las columnas nombre y apellido que acabamos de crear y llenar
```

```python
#Consigna 3: Obtener medidas de resumen del conjunto de datos. Cuál es el país que ganó más medallas?
resumen_consignas=consignas.describe() #creamos el resumen del conjunto de datos
resumen_consignas.loc["top"]["Country"] #con esta linea veremos el país que más ganó medallas el cual es United States
```

```python
#Consigna 4: Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizo el evento.

medallasporhombres=consignas[consignas.Gender=="Men"] #aqui asignamos el filtro de "Men" a medallasporhombres
medallasporhombres.groupby(["Gender", "Year"]).size() #con esta linea procederemos a agrupar por el genero y el año, para ver cuantas medallas ganaron en ciertos años
```
